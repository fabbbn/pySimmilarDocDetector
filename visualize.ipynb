```python
ids = np.array(exc_time.groupby(by=['doc_id']).nunique().reset_index()['doc_id'])
manning = np.array(frame.loc[frame['config_used']=='manning']['tot_retrieved'])
jiffriya = np.array(frame.loc[frame['config_used']=='jiffriya']['tot_retrieved'])
xu = np.array(frame.loc[frame['config_used']=='xu']['tot_retrieved'])
saptono = np.array(frame.loc[frame['config_used']=='saptono']['tot_retrieved'])

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(ids, manning, label='Manning')
ax.plot(ids, jiffriya, label='Jiffriya')
ax.plot(ids, xu, label='Xu')
ax.plot(ids, saptono, label='Saptono')
# ax.set_xlabel('Kode Dokumen')
ax.set_xlabel('Document Number')
ax.set_ylabel('Number of Retrieved Documents')
# ax.set_ylabel('Jumlah Bagian Dokumen Retrieved')
ax.set_title('Number of Retrieved Documents Comparison Based on Configuration')
# ax.set_title('Perbandingan Jumlah Bagian Dokumen Retrieved')

ax.legend()
```

```python
ids = np.array(exc_time.groupby(by=['doc_id']).nunique().reset_index()['doc_id'])
manning = np.array(frame.loc[frame['config_used']=='manning']['highest_cosine'])
jiffriya = np.array(frame.loc[frame['config_used']=='jiffriya']['highest_cosine'])
xu = np.array(frame.loc[frame['config_used']=='xu']['highest_cosine'])
saptono = np.array(frame.loc[frame['config_used']=='saptono']['highest_cosine'])
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(ids, manning, label='Manning')
ax.plot(ids, jiffriya, label='Jiffriya')
ax.plot(ids, xu, label='Xu')
ax.plot(ids, saptono, label='Saptono')
# ax.set_xlabel('Kode Dokumen')
ax.set_xlabel('Document Number')
# ax.set_ylabel('Nilai Cosine Similarity')
ax.set_ylabel('Cosine Similarity Value')
# ax.set_title('Perbandingan Nilai Cosine Similarity Tertinggi antar Dokumen')
ax.set_title('Document\'s Highest Cosine Similarity Value Comparison Based on Configuration')
ax.legend()
```
